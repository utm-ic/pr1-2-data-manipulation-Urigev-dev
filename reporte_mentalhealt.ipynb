## Paso 3 Limpieza de datos

```python
null_counts = df.isna().sum()
null_percentage = (null_counts / len(df)) * 100
print("Porcentaje de valores nulos por columna:\n", null_percentage)
```

### ¿Qué porcentaje del dataset contiene valores nulos?

Medication_Use: 62.25%

Substance_Use: 69.50%

Las demás columnas no presentan valores nulos.

Esto representa que 2 de 21 columnas (9.5%) tienen datos faltantes, pero afectan a más del 60% de los registros.

```python
# Verificar duplicados
duplicate_count = df.duplicated().sum()
print(f"Número de registros duplicados: {duplicate_count}")
```

```python
# Agregar la categoría 'Desconocido' antes de rellenar nulos
for col in ['Medication_Use', 'Substance_Use']:
    df[col] = df[col].astype('category')
    if 'Desconocido' not in df[col].cat.categories:
        df[col] = df[col].cat.add_categories(['Desconocido'])
    df[col] = df[col].fillna('Desconocido')

# Convertir otras columnas categóricas
categorical_columns = ['Gender', 'Education_Level', 'Employment_Status']
df[categorical_columns] = df[categorical_columns].astype('category')

# Eliminar duplicados
df_cleaned = df.drop_duplicates()

# Confirmación final
print("¿Existen valores nulos tras la limpieza?:", df_cleaned.isnull().sum().sum() > 0)
print(f"Total de registros tras limpieza: {len(df_cleaned)}")
```

### ¿Qué estrategia se aplicó para tratar los valores nulos y por qué?

Se optó por rellenar los valores nulos con la categoría 'Desconocido', ya que ambas columnas (Medication_Use y Substance_Use) son variables categóricas.
Esta decisión evita eliminar registros valiosos (más del 60% del dataset) y permite capturar la ausencia de respuesta como un posible patrón relevante.

## Paso 4: Análisis Estadístico

```python

# Seleccionar las columnas a analizar
columns_to_analyze = ['Anxiety_Score', 'Depression_Score', 'Stress_Level']

# Calcular media, mediana y desviación estándar
statistics = df_cleaned[columns_to_analyze].agg(['mean', 'median', 'std']).T

# Renombrar columnas para presentación
statistics = statistics.rename(columns={
    'mean': 'Media',
    'median': 'Mediana',
    'std': 'Desviación Estándar'
})

# Mostrar resultados
print("Estadísticas descriptivas de puntajes psicológicos:")
print(statistics)
```

### ¿Cuál de los tres puntajes tiene mayor variabilidad?

Anxiety_Score es el puntaje con mayor variabilidad, ya que presenta la desviación estándar más alta: 5.91

### ¿Qué observas sobre la diferencia entre media y mediana en cada caso? ¿Qué te dice esto sobre la distribución?

Cuando la media y la mediana son similares, como en Stress_Level y Anxiety_Score, se trata de una distribución aproximadamente normal o simétrica.

En Depression_Score, la mediana es ligeramente superior a la media, lo que sugiere una leve asimetría negativa (sesgo a la izquierda), es decir, algunos valores bajos están “empujando” la media hacia abajo.

## Paso 5: Filtrado de Datos

```python
# Puntaje alto de ansiedad (>80% del máximo)
puntaje_ansiedad = df['Anxiety_Score'].max() * 0.8

# Personas con alta ansiedad
puntaje_ansiedad_df = df[df['Anxiety_Score'] > puntaje_ansiedad]

# Porcentaje con ansiedad alta
porc_ans_alto = (len(puntaje_ansiedad_df) / len(df)) * 100
print(f"Ansiedad alta: {porc_ans_alto:.2f}%")

# Categorizar horas de sueño
bins = [0, 5, 7, df['Sleep_Hours'].max()]
labels = ['<5 hrs', '5-7 hrs', '>7 hrs']
df['Cat_Sueño'] = pd.cut(df['Sleep_Hours'], bins=bins, labels=labels, include_lowest=True)

# Estrés promedio por categoría de sueño
estres_sueño = df.groupby('Cat_Sueño', observed=True)['Stress_Level'].mean()
print("\nEstrés promedio según horas de sueño:")
print(estres_sueño)
```

### ¿Qué porcentaje del total presenta niveles altos de ansiedad?

El 21.0% del total del dataset presenta niveles altos de ansiedad, considerando como nivel alto un puntaje por encima de 16 (que representa más del 80% del puntaje máximo).

### ¿Existe una relación visible entre las horas de sueño y el nivel de estrés?

Sí, existe una relación visible, aunque leve, entre las horas de sueño y el nivel promedio de estrés. Quienes duermen menos de 5 horas tienden a experimentar niveles ligeramente más elevados de estrés, mientras que dormir entre 5 y 7 horas se asocia con niveles ligeramente menores.

## Paso 7: Guardado de Resultados

```python
# Calcular la mediana de estrés
mediana_estres = df['Stress_Level'].median()

# Filtrar personas con terapia y estrés bajo
terapia_bajo_estres = df[(df['Therapy'] == 1) & (df['Stress_Level'] < mediana_estres)]

# Número total de personas
cantidad = len(terapia_bajo_estres)
print(f"Total de personas con terapia y estrés bajo: {cantidad}")

# Guardar resultado en CSV
terapia_bajo_estres.to_csv('personas_terapia_bajo_estres.csv', index=False)
```

### ¿Cuántas personas cumplen con ambas condiciones?

108 personas reciben terapia y tienen un nivel bajo de estrés (por debajo de la mediana).

### ¿Qué conclusiones puedes hacer sobre la relación entre terapia y nivel de estrés en este conjunto?

Un número significativo de personas que reciben terapia muestra niveles reducidos de estrés, lo que puede indicar que la terapia tiene una relación positiva con la reducción o manejo efectivo del estrés.

Estos datos podrían apoyar la hipótesis de que la terapia contribuye a un mejor manejo emocional y psicológico, reflejado en menores niveles reportados de estrés.

## Paso 8: Visualización Opcional

```python
# Calcular la media de autoestima según género
autoestima_por_genero = df.groupby('Gender', observed=True)['Self_Esteem_Score'].mean()

# Crear gráfico de barras
plt.figure(figsize=(8,5))
autoestima_por_genero.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Media de Autoestima por Género')
plt.xlabel('Género')
plt.ylabel('Puntaje Promedio de Autoestima')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.tight_layout()
plt.show()

autoestima_por_genero
```

### ¿Hay diferencias notables entre los géneros en cuanto a autoestima?

Sí, existen diferencias claras:

"Female" y "Male" presentan puntajes de autoestima similares, con ligera ventaja para "Female".

"Non-Binary" y especialmente "Other" presentan una autoestima notablemente más baja.

### ¿Qué factores del dataset podrían explicar estas diferencias?

Los factores del dataset que podrían explicar estas diferencias incluyen:

Apoyo social: Personas con menos soporte social suelen mostrar menor autoestima.

Nivel de estrés y ansiedad: Niveles elevados pueden afectar negativamente la autoestima.

Discriminación o exclusión percibida: Podría ser relevante para los grupos con etiquetas "Non-Binary" u "Other".

Nivel de satisfacción con la vida y soledad percibida: Personas con mayor satisfacción suelen mostrar mejor autoestima.
